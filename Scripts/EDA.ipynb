reassignment_method, reassignment_reason, cancelled_time consists of 99% null values because If an order goes smoothly, it is assigned to one rider who picks it up and delivers it. It never gets reassigned. Therefore, those columns should be empty 99% of the time. That means That 0.1% of data where those columns are filled represents your operational failures.

If delivered_time or pickup_time is missing (NULL), it almost always means the order was cancelled or undelivered. We will leave them as Null which will tell us that delivery chain broke down.

If a rider has NULL for lifetime_order_count, it doesn't mean they are an "average" rider. It usually means this is their very first day on the job. We will fill that with 0 instead of mean.
