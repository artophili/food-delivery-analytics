```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("C:/Users/Artophilic/Analysis Projects/Swiggy_order_analysis/Data/Rider-Info.csv")
df.head()
```

```python
#Converting the time columns to datetime format

time_columns = ['order_time', 'order_date', 'allot_time', 'accept_time', 'pickup_time', 'delivered_time', 'cancelled_time']

for col in time_columns:
    df[col] = pd.to_datetime(df[col], errors='coerce')
```

```python
#Filling Reassignment columns
df['reassignment_method'] = df['reassignment_method'].fillna('No Reassignment')
df['reassignment_reason'] = df['reassignment_reason'].fillna('No Reassignment')
```

```python
df['reassigned_order'] = np.where(df['reassignment_method'] != 'No Reassignment', 1, 0)
```

```python
#rider baseline performance columns
rider_metrics = ['alloted_orders', 'delivered_orders', 'undelivered_orders', 'lifetime_order_count']
for metric in rider_metrics:
    df[metric] = df[metric].fillna(0).astype(int)
```

```python
#Generate a unique composite key for the Geography Dimension
df['location_id'] = df.groupby(['first_mile_distance', 'last_mile_distance']).ngroup() + 1
```

```python
df.head()
```

```python
df['session_time'] = df['session_time'].fillna(df['session_time'].median())
```
